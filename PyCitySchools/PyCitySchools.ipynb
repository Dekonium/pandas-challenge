```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left")
```

```python
school_data_complete.head()
```

```python
total_schools = school_data_complete["school_name"].nunique()

total_students = school_data_complete["student_name"].count()

total_budget = school_data["budget"].sum()


av_math = school_data_complete["math_score"].mean()

av_read = school_data_complete["reading_score"].mean()


pass_math = pd.DataFrame((school_data_complete.loc[school_data_complete["math_score"]>69]))
countmath = pass_math["math_score"].count()


pass_read = pd.DataFrame((school_data_complete.loc[school_data_complete["reading_score"]>69]))
countread = pass_read["reading_score"].count()

pass_both = student_data.loc[(school_data_complete["math_score"]>69) & (school_data_complete["reading_score"]>69)]
bothcount = pass_both.shape[0]



perc_math = ((countmath/total_students)*100)

perc_read = ((countread/total_students)*100)

perc_both = ((bothcount/total_students)*100)




columns = ["Total Schools","Total Students","Total Budget","Average Math Score",\
           "Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]
rows = [total_schools,total_students,total_budget,av_math,av_read,perc_math,perc_read,perc_both]
print(columns)
print(rows)
summary = dict(zip(columns,rows))
summary
summary_df = pd.DataFrame(summary,index=[0])
summary_df["Total Budget"] = summary_df["Total Budget"].map("${:,.2f}".format)
summary_df
```

```python
school_data_complete["budget_per_student"] = (school_data_complete["budget"]/school_data_complete["size"])
school_data_complete["Passed_math"] = school_data_complete["math_score"] > 69
school_data_complete["Passed_read"] = school_data_complete["reading_score"] > 69
school_data_complete["Passed_both"] = (school_data_complete["math_score"] > 69) & (school_data_complete["reading_score"] > 69)


school_data_complete_group = school_data_complete.groupby(["school_name","type"])
school_data_complete_avg = school_data_complete_group.mean()

school_data_complete_avg = school_data_complete_avg.drop("Student ID",axis = "columns")
school_data_complete_avg = school_data_complete_avg.drop("School ID",axis = "columns")
school_data_complete_avg["Passed_math"]=school_data_complete_avg["Passed_math"]*100
school_data_complete_avg["Passed_read"]=school_data_complete_avg["Passed_read"]*100
school_data_complete_avg["Passed_both"]=school_data_complete_avg["Passed_both"]*100

summary_schools = school_data_complete_avg.rename(columns={"Passed_math":"%Passed_math","Passed_read":"%Passed_read",\
                                                           "Passed_both":"%Passed_both","size":"Total_students"})
summary_schools["budget"] = summary_schools["budget"].map("${:,.2f}".format)
#summary_schools["budget_per_student"] = summary_schools["budget_per_student"].map("${:,.2f}".format)

summary_schools

```
